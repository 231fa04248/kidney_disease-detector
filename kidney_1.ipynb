```python
import pandas as pd
df=pd.read_csv('/content/kidney_disease.csv')
df_cleaned=df.dropna()
df_cleaned.to_csv('cleaned_file.csv',index=False)
```

```python
import pandas as pd
df = pd.read_csv('cleaned_file.csv')
rbc_pc_mapping = {'normal': 1, 'abnormal': 0}
appet_mapping = {'poor': 0, 'good': 1}
pcc_ba_mapping = {'present': 1, 'notpresent': 0}
classification_mapping = {'ckd': 1, 'notckd': 0}
if 'rbc' in df.columns:
    df['rbc'] = df['rbc'].map(rbc_pc_mapping)
if 'pc' in df.columns:
    df['pc'] = df['pc'].map(rbc_pc_mapping)
if 'appet' in df.columns:
    df['appet'] = df['appet'].map(appet_mapping)
if 'pcc' in df.columns:
    df['pcc'] = df['pcc'].map(pcc_ba_mapping)
if 'ba' in df.columns:
    df['ba'] = df['ba'].map(pcc_ba_mapping)
if 'classification' in df.columns:
    df['classification'] = df['classification'].map(classification_mapping)
yes_no_columns = ['htn', 'dm', 'cad', 'pe', 'ane']
for col in yes_no_columns:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].map({'yes': 1, 'no': 0})
df.to_csv('transformed_file.csv', index=False)

print("Transformation complete. Data saved to 'transformed_file.csv'")
```

```python
import keras   #Imports
```

```python
from keras.models import Sequential
obj = Sequential()   #Model Creation
```

```python
from keras.layers import Dense
```

```python
import pandas as pd

dt = pd.read_csv('/content/transformed_file.csv')   #Data Loading
dt.head()
```

```python
dt.corr()
```

```python
dt = dt.drop(['pcc','ba'],axis=1)
```

```python
dt.tail(1)
```

```python
x = dt.iloc[:,0:11]   #Feature Selection
y = dt['classification']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split   #Data Split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
obj.add(Dense(units=11,activation='relu'))  #Model Architecture
```

```python
obj.add(Dense(units=5,activation='relu'))
```

```python
obj.add(Dense(units=1,activation='sigmoid'))
```

```python
obj.build(input_shape=(None,11))
```

```python
obj.summary()
```

```python
obj.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
obj.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
y_pred = obj.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score  #Evaluation
accuracy_score(y_test,y_pred)
```

```python
obj.save('kidney.h5')
```
